### Import Required labraries 

```python
import pandas as pd 
import numpy as np 
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import cross_val_score
import joblib
```

### Load Processed Data 

```python
train_df=pd.read_csv("C:\\Users\\JOHN\\Desktop\\ML_projects\\crop-recommendation-system\\data\\processed\\train_data.csv")
test_df=pd.read_csv("C:\\Users\\JOHN\\Desktop\\ML_projects\\crop-recommendation-system\\data\\processed\\test_data.csv")
```

```python
X_train=train_df.drop('label',axis=1)
y_train=train_df['label']
```

```python
X_test=test_df.drop('label',axis=1)
y_test=test_df['label']
```

### Baseline model :: Logestic regression 

```python
lr=LogisticRegression(max_iter=1000)
```

```python
lr.fit(X_train,y_train)
```

```python
lr_pred=lr.predict(X_test)
```

```python
lr_acuracy=accuracy_score(y_test,lr_pred)
print("Logistic regression Acuracy score ::",lr_acuracy)
```

### Advanced Model :: Random forest 

```python
rf=RandomForestClassifier(
    n_estimators=200,
    random_state=42,
    n_jobs=-1
)
```

```python
rf.fit(X_train,y_train)
```

```python
rf_pred=rf.predict(X_test)
```

```python
rf_acuracy=accuracy_score(y_test,rf_pred)
print("Random forest Acuracy score ::",rf_acuracy)
```

### Cross-Validation (Model Stability)

```python
rf_cv=cross_val_score(rf,X_train,y_train,cv=10,scoring='accuracy')
```

```python
print("RF CV Mean Accuracy:", rf_cv.mean())
print("RF CV Std:", rf_cv.std())
```

### Detailed Evaluation 

```python
print(classification_report(y_test,rf_pred))
```
